# Her2

```python
subtype = 'Her2'
common_all_seeds = []

for seed in (list(range(5)) + [55, 666, 7777, 88888, 999999]):
  elrp = pd.read_csv('Attributions/seed' + str(seed) + '/elrp_' + subtype + '_pos.csv')
  deeplift = pd.read_csv('Attributions/seed' + str(seed) + '/deeplift_' + subtype+ '_pos.csv')
  grad = pd.read_csv('Attributions/seed' + str(seed) + '/grad*input_' + subtype+ '_pos.csv')
  intgrad = pd.read_csv('Attributions/seed' + str(seed) + '/intgrad_' + subtype+ '_pos.csv')
  gradshap = pd.read_csv('Attributions/seed' + str(seed) + '/grad_shap_' + subtype+ '_pos.csv')
  deepshap = pd.read_csv('Attributions/seed' + str(seed) + '/deep_shap_' + subtype+ '_pos.csv')

  elrp = elrp[elrp['Frequency'] >= round(0.25*31)]
  deeplift = deeplift[deeplift['Frequency'] >= round(0.25*31)]
  grad = grad[grad['Frequency'] >= round(0.25*31)]
  intgrad = intgrad[intgrad['Frequency'] >= round(0.25*31)]
  gradshap = gradshap[gradshap['Frequency'] >= round(0.25*31)]
  deepshap = deepshap[deepshap['Frequency'] >= round(0.25*31)]
  

  methods = [elrp, deeplift, grad, intgrad, gradshap, deepshap]
  common = []
  for method in methods:
    common += list(set(method[subtype+'Genes']))
  common_pos = dict(Counter(common))

  print(len(list(set([key for key, value in common_pos.items() if value >= 5]))))
  common_all_seeds += list(set([key for key, value in common_pos.items() if value >= 5]))
 
```

```python
her2_genes = dict(Counter(common_all_seeds))
her2_freq = dict(sorted(her2_genes.items(), key=lambda item: item[1], reverse=True))
her2_freq_df = pd.DataFrame(her2_freq.items(), columns=[subtype+'Genes', 'Frequency'])
```

```python
her2_freq_df[her2_freq_df['Frequency']>=seed_cutoff]
```
